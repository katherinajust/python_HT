```python
import pandas as pd
```

```python
#Задание1
ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
```

```python
df = ratings.merge(movies, on='movieId', how='left')
df.head()
```

```python
df = df.filter(items = ['title', 'rating'])
df.head()
```

```python
new_df = df[df['rating'] == 5.0]
#new_df
new_df.groupby('title').count().sort_values('rating', ascending=False).head()
```

```python
#Задание2
power = pd.read_csv('power.csv')
power.head()
```

```python
new_power = power[(((power['country']=='Latvia') | (power['country']=='Lithuania') | (power['country']=='Estonia')) & ((power['category'] == 4) | (power['category'] == 12)  | (power['category'] == 21)) & ((power['year'] >= 2005) & (power['year'] <= 2010)) & (power['quantity'] >= 0))]
result = new_power.filter(items = ['country', 'quantity']).groupby('country').sum()
result
```
