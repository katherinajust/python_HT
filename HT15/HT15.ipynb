```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import json
```

```python
KEYWORDS = ['технологии', 'маркетинг', 'блокчейн']
```

```python
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
articles = soup.find_all('article', class_ = 'post post_preview')
articles
```

```python
article_list = []
for el in articles:
    #print(el)
    if KEYWORDS[0] in str(el) or KEYWORDS[1] or KEYWORDS[2] in str(el):
        article_list.append(el)
print (article_list)
```

```python
article_date = list(map(lambda x: x.find('span', class_ = 'post__time'), article_list))
article_title = list(map(lambda x: x.find('h2', class_ = 'post__title').get_text(), article_list))
article_link = list(map(lambda x: x.find('a').get('href'), article_list))
```

```python
df = pd.DataFrame({'Дата': article_date, 'Название':article_title, 'Ссылка':article_link })
df
```

```python
#Задание 2
URL = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
headers = {'vaar-header-app-product':'hackcheck-web-avast',
           'vaar-version': '0'}
body = {'emailAddresses': ['nechay123@ya.ru', 'e.nechay@gos-fin.ru']}
```

```python
s = requests.post(URL, headers = headers, data=json.dumps(body))
s.json()
```

```python
def check(new_df):
    for keys, values in s.json()['data'].items():
        for val in values:
            if new_df == keys:
                return val
```

```python
df_breaches_a = pd.DataFrame({'Ресурс':s.json()['breaches']['10']['site'], 'Дата':s.json()['breaches']['10']['publishDate'], 'Описание':s.json()['breaches']['10']['description']}, index = [0])
df_breaches_b = pd.DataFrame({'Ресурс':s.json()['breaches']['16768']['site'], 'Дата':s.json()['breaches']['16768']['publishDate'], 'Описание':s.json()['breaches']['16768']['description']}, index = [0])
df_breaches_c = pd.DataFrame({'Ресурс':s.json()['breaches']['4458']['site'], 'Дата':s.json()['breaches']['4458']['publishDate'], 'Описание':s.json()['breaches']['4458']['description']}, index = [0])
new_df = pd.concat([df_breaches_a, df_breaches_b, df_breaches_c ], ignore_index=True)
new_df['Почта'] = new_df['Ресурс'].apply(check)
new_df
```
