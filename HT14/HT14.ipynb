```python
#Задание 1
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings_df = ratings.groupby('userId').filter(lambda x: x['rating'].count()>100)
```

```python
ratings_df = ratings_df[['userId', 'timestamp']]
df = ratings_df.groupby('userId').agg([min, max])
df['diff'] = df['timestamp']['max'] - df['timestamp']['min']
df['diff'].mean() #/24/3600
```

```python
from datetime import datetime, timedelta
str(timedelta(seconds=int(df['diff'].mean())))
```
